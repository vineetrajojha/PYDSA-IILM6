```python
#selection sort

#function

#take length of arr in a variable

#for loop

#take a minimum index

#for j loop

#if < minimum index

#minimum index = j

#swap arr i

def ss(arr):
  v = len(arr)
  for i in range(v):
    minimum_index = i
    for j in range(i+1, v):
      if arr[j] < arr[minimum_index]:
        minimum_index = j

    arr[i], arr[minimum_index] = arr[minimum_index], arr[i]
  return arr

l1=[1,6,2,3,7,4]
print(f"before sort: {l1}")
sorted_arr = ss(l1)
print(f"selection sorted: {sorted_arr}")

```

```python
#merge sort

#function ms
#if length of arr > 1
#mid = len(arr)//2
#before arr[:mid] = l and after mid = r

#ms(L) and (R)
#i,j,k,0 =

def ms(arr):
  if len(arr) > 1:
    mid = len(arr)//2

    L = arr[:mid]
    R = arr[mid:]

    i = j = k = 0

    while i < len(L) and j < len(R):
      if L[i] < R[j]:
        arr[k] = L[i]
        i+=1
      else:
        arr[k] = R[i]
        j+=1
      k += 1

    while i < len(L):
      arr[k] = L[i]
      i += 1
      k += 1


    while j < len(R):
      arr[k] = R[j]
      j += 1
      k += 1


l1=[1,4,5,6,2,3]
print(f"before sort : {l1}")
sorted_arr=ms(l1)
print(f"sorted array : {sorted_arr}")

```

```python
#bubble sort

def bs(arr):
  v = len(arr)


  for i in range(v):
    change_position = False
    for j in range(0,v-i-1):
      if arr[j] > arr[j+1]:
        arr[j], arr[j+1] = arr[j+1], arr[j]
        change_position = True

    if not change_position:
      break
  return arr


l1=[1,4,5,6,2,3]
print(f"before sort : {l1}")
sorted_arr=bs(l1)
print(f"sorted array : {sorted_arr}")
```

```python
#design a function to check pangram programme

import string

def pangram(sentence):
  alphabet = set(string.ascii_lowercase)

  letters = set(char.lower() for char in sentence if char.isalpha())

  return alphabet <= letters

s1 = str(input("enter a sentence"))

print(f"'{s1}' is a pangram: {pangram(s1)}")
```

```python
#program that defines the function that counts_alphabet_digit() that accepts a string and calcute the number of strings init

def counts_alphabet_digit(s):
    alphabets = 0
    digits = 0
    for char in s:
        if char.isalpha():
            alphabets += 1
        elif char.isdigit():
            digits += 1
    return alphabets, digits

text = input("Enter a string: ")
a, d = counts_alphabet_digit(text)
print(f"Alphabets: {a}")
print(f"Digits: {d}")
```

```python
#sum elemets of a list recursively ( asume homogenous lists)

#min/max of a list recursively

a = [1,2,3,4,5]

def listSummer(a: list[int], ind : int):
  #base case
  if ind == len(a) : return 0
  #recursion case
  return a[ind] + listSummer(a, ind+1)

def minElement(a : list[int], ind : int):
  #base case
  if ind == len(a) : return 0
  #rec case
  minAhead = minElement(a, ind+1)
  return min(a[ind], minAhead)

def maxElement(a : list[int], ind : int):
  #base case
  if ind == len(a) : return 0
  #rec case
  maxAhead = maxElement(a, ind+1)
  return max(a[ind], maxAhead)

print(listSummer(a,0))
print(minElement(a, 0))
```

```python
#check whether a string is palindrome or not ( recursively)
def is_palindrome_str(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome_str(s[1:-1])

s_input = input("Enter a string: ")
print(f"Is '{s_input}' a palindrome? {is_palindrome_str(s_input)}")

#check whether a number is palindrome or not
def is_palindrome_num(n):
    s = str(n)
    return is_palindrome_str(s)

n_input = int(input("Enter a number: "))
print(f"Is {n_input} a palindrome? {is_palindrome_num(n_input)}")
```
